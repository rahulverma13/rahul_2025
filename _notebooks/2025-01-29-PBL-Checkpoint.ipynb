## Progress Updates: What has changed?

For our initial feature, we hadn't thought through the entire process of how the feature would work. For students to be able to grade assignments, a teacher must be able to "assign" certain students to be "graders" for certain assignments.

This caused large changes to the code structure. I created a many to many relationship on the backend, connecting students to assignments if the teacher assigns them. See here:

### 2. Unification/Connection of AssignmentSubmission & SynergyGrades database values

Before the last 2 weeks, there was a disconnect between ``AssignmentSubmission`` and ``SynergyGrades``. When the teacher would grade an assignment, it would get connected to the submission, but not in the grades because they are seperate. To fix this, I worked on connecting the AssignmentSubmission feature to additionally save to SynergyGrades. This applies when the teacher grades the assignment in student or teacher mode.

We considered consolidating these two classes into one, but decided not too because there may be many submissions, but only 1 grade for the assignment. Thus, this opted as the simplest & most effective option to avoid this disconnect between Grades and Submissions.

### 3. Connecting Student Comments to Teacher Grading
In collaboration with Kayden, I connected his comments submission to showing up on the teacher grading assignments page. With the teacher also being able to assign feedback, this creates a multifaceted comment system: students can submit comments with their submission, and then the teacher can submit comments back.
